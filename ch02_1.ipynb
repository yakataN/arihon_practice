CPPのコード共有したい場合はbatcatで表示する

```python
folder_name = "ch02_1/"
files = !ls ./ch02_1
cpps = [i for i in files if i[len(i)-3:]=='cpp']
inputs = [i for i in files if i[len(i)-5:]=='input']
print(cpps)
print(inputs)
```

```python
def matome(index, input_size = None):
    file_name = cpps[index-1]
    core = folder_name + file_name[:-4]
    cpp = core + ".cpp"
    out = core + ".out"
    !g++ $cpp -o $out
    !batcat $cpp
    if (input_size==None):
        !./$out | nl -w2 -nrz -s": " 
    else:
        for i in range(input_size):
            # input = "{}_{}.input".format(core, i+1)
            input_file = "./" + core + "_" + str(i+1) + ".input"
            print(input_file)
            !./$out < $input_file | nl -w2 -nrz -s": " 
            print()
```
